```python
import csv
import pandas as pd
from datetime import datetime

df = pd.read_csv(r"C:\Users\Акмира\OneDrive\Документы\GitHub\ICT_2022\practice1\MRNA.csv", parse_dates=["Date"])

df['Name'] = 'Akmira'
df['Surname'] = 'Sagatbek'
df['Date_of_download'] = datetime(year = 2022, month = 9, day = 8)

percentage_change = [0]
for i, row in df.iterrows():
    
    if i == 0: 
        continue
    else:
        if df['Open'][i] != 0:
            percentage_change.append((df['Open'][i] - df['Open'][i-1])/df['Open'][i] * 100)
        else: 
            percentage_change.append(0)
df['percentage_change'] = percentage_change

print(df)
```
