```python
import numpy as np
import sklearn.linear_model as skmod
```

```python
alt = [3.25, 4.2, 1.9]
n_goats = [21, 22, 13]

arr_x = np.array(alt)
arr_x = arr_x.reshape(-1, 1)

arr_y = np.array(n_goats).reshape(-1, 1)

print(arr_x)
print(arr_y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(arr_x, arr_y)
```

```python
model = skmod.LinearRegression()
```

```python
model.trained = model.fit(arr_x, arr_y)
```

```python
print(model.trained.coef_)
print(model.trained.intercept_)

print(model.trained.score(arr_x, arr_y))
```

```python
plt.scatter(arr_x, arr_y)
plt.plot([0, 5], [-1.83*0 + 26.6, -1.83 * 5 + 26.6])
```

```python
x1 = [9, 10, 16, 22, 17, 7, 10, 24, 15]
x2 = [14, 18, 1, 22, 6, 12, 4, 28, 22]
y = [42.8, 48.8, 36.8, 81.1, 44.8, 32.4, 28.8, 100, 73.1]

plt.scatter(x1, x2, c = y, cmap = 'Reds')

ax = plt.subplot(projection='3d')
ax.scatter(x1, x2, y)
```

```python
arr_x1 = np.array(x1).reshape(-1, 1)
arr_x2 = np.array(x2).reshape(-1, 1)

arr_y = np.array(y).reshape(-1, 1)

arr_x = np.hstack([arr_x1, arr_x2])
print(arr_x)
print(arr_y)
```

```python
print(model.coef_)
print(model.intercept_)
#2.213 * x1 + 1.689 * x2 - 1.507
```

```python
alt = [3.25, 0.816, 4.376, 1.314, 3.982, 2.957, 2.482, 3.7]
n_goats = [21, 22, 13, 25, 17, 23, 23, 27]

alt_sq = []
for i in alt:
    alt_sq.append(i ** 2)

arr_x1 = np.array(alt).reshape(-1, 1)
arr_x2 = np.array(alt_sq).reshape(-1, 1)

arr_y = np.array(n_goats).reshape(-1, 1)

arr_x = np.hstack([arr_x1, arr_x2])
print(arr_x)
print(arr_y)
```

```python
model = skmod.LinearRegression().fit(arr_x, arr_y)
```

```python
print(model.coef_)
print(model.intercept_)
#8.23 * x1 + -1.97 * x2 + 16.66
```

```python
plt.scatter(arr_x1, arr_y)

x_for_plot = np.arange(0, 5, 0.1)
x2_for_plot = x_for_plot * x_for_plot
x_final = np.hstack([x_for_plot.reshape(-1, 1), x2_for_plot.reshape(-1, 1)])
y_for_plot = model.predict(x_final)

plt.plot(x_for_plot, y_for_plot)
# print(x2_for_plot)
```

```python
x = [13, 3, 17, 0, 2]
y = [505, 35, 836, -6, 16]

alt_sq = []
for i in x:
    alt_sq.append(i ** 2)

arr_x1 = np.array(alt_sq).reshape(-1, 1)
arr_x2 = np.array(x).reshape(-1, 1)

arr_y = np.array(y).reshape(-1, 1)

arr_x = np.hstack([arr_x1, arr_x2])
print(arr_x)
print(arr_y)
```
