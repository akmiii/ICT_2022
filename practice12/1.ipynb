```python
import pandas as pd

df = pd.read_parquet(r"C:\Users\Акмира\OneDrive\Документы\GitHub\ICT_2022\data_streamers.parquet")
print(df)
```

```python
import numpy as np

x1 = df['Time_of_the_day'].to_numpy().reshape(-1, 1)
x2 = df['Number_of_viewers'].to_numpy().reshape(-1, 1)
x3 = df['Average_age_of_viewers'].to_numpy().reshape(-1, 1)

y = df['Video_game'].to_numpy().ravel()
print(x1)
print(x2)
print(x3)
print(y)
```

```python
import sklearn.model_selection as sksel
import sklearn.preprocessing as skprepro
import sklearn.linear_model as skmod

encoder = skprepro.LabelEncoder()
encoder = encoder.fit(y)
fav_code = encoder.transform(y)

print(fav_code)
print(encoder.classes_)
```

```python

x = np.hstack([x1, x2, x3])

model = skmod.LogisticRegression(multi_class = 'ovr')
model = model.fit(x, fav_code)

print(model.score(x, fav_code))
print(x)
```

```python
print(model.predict_proba(np.array([[   11 ,10000 ,   18]])))
print(model.predict_proba(np.array([[   23 ,1000 ,   30]])))

print(encoder.classes_)

print(encoder.inverse_transform(model.predict(np.array([[   11 ,10000 ,   18]]))))
print(encoder.inverse_transform(model.predict(np.array([[   23 ,1000 ,   30]]))))
# print(encoder.inverse_transform(model.predict(np.array([[20], [30], [40]]))))
# print(model.predict_proba(np.array([[20], [30], [40]])))
```
