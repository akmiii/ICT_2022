```python
import pandas as pd
import csv
from datetime import datetime
from time import strptime

with open("MRNA.csv", "r") as file:
    csv_file = csv.reader(file)
    date = []
    open1 = []
    high = []
    low = []
    close = []
    adj_close = []
    volume = []
    for row in csv_file:
        date.append(row[0])
        open1.append(row[1])
        high.append(row[2])
        low.append(row[3])
        close.append(row[4])
        adj_close.append(row[5])
        volume.append(row[6])
new_date, new_open, new_high, new_low, new_close, new_adj, new_volume = [], [], [], [], [], [], []


for i in range(1, len(date)):
    new_date.append(datetime.strptime(date[i], "%Y-%m-%d"))
    new_open.append(float(open1[i]))
    new_high.append(float(high[i]))
    new_low.append(float(low[i]))
    new_close.append(float(close[i]))
    new_adj.append(float(adj_close[i]))
    new_volume.append(float(volume[i]))

dict = {}

dict = {'Date' : new_date,'Open' : new_open, 'High' : new_high, 'Low' : new_low, 'Close' : new_close, 'Adj Close' : new_adj, 'Volume' : new_volume}

# print(dict)

dt = pd.DataFrame(dict)
dt.columns = ['New Date', 'New Open', 'High Price', 'Low Price', 'New Close', 'New Adj Close', 'New Volume']
dt['Name'] = ['Akmira' for _ in range(944)]
dt['Surname'] = ['Sagatbek' for _ in range(944)]
dt['Date of download'] = [datetime(year = 2022, month = 9, day = 8) for _ in range(944)]

date_begin_mean = datetime(year = 2021, month = 6, day = 1)
date_end_mean = datetime(year = 2021, month = 12, day = 31)

dt = dt[(dt['New Date'] >= date_begin_mean) & (dt['New Date'] <= date_end_mean)] 
print(dt)
```
